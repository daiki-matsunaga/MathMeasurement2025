```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150

# 可視化用：x, yのデータ生成
grid = np.linspace(-1, 1, 31)
lx, ly = np.meshgrid(grid, grid)
lx = lx.flatten()
ly = ly.flatten()
```

```python
def Compute(w_ij):
    lz = np.array([])

    # zjを取り出しziを計算
    for x, y in zip(lx, ly):
        zj = np.array([x, y])
        zi = 1.0/(1.0 + np.exp(-w_ij@zj))

        lz = np.append(lz, zi)

    return lz

def ShowFigure(lz):
    # グラフに表示
    fig, ax = plt.subplots()
    sc = ax.scatter(lx, ly, c=lz, vmax=1.0, vmin=0.0)
    ax.set_aspect('equal')
    fig.colorbar(sc, ax=ax)
    plt.show()
```

```python
# 初期にランダムな重み行列を生成
w_ij = np.random.normal(0.0, 1.0, (1, 2))
print(f'w_ij =\n {w_ij}')

# 乱数のw_ijでマップを生成する
zi = Compute(w_ij)
ShowFigure(zi)
```

(a) $x>0$で出力が1，そうでなければ0となるような重みベクトルを手入力で実装せよ

```python
# 実装
w_ij = np.array([0.0, 0.0])

# マップを生成する
zj = Compute(w_ij)
ShowFigure(zj)
```

(b) $y<0$で出力が1，そうでなければ0となるような重みベクトルを手入力で実装せよ

```python
# 実装
w_ij = np.array([0.0, 0.0])

# マップを生成する
zj = Compute(w_ij)
ShowFigure(zj)
```
